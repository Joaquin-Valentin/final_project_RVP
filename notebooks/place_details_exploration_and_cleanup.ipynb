```python
google_imputed['result.price_level'].value_counts()
```

```python
def round_decimals(x):
    if x.is_integer():
        return x
    else:
        return int(x) + round(x - int(x))
```

```python
google_imputed['result.price_level'] = google_imputed['result.price_level'].apply(round_decimals)
```

```python
import scipy.stats as st
```

```python
def t_test_features(s1, s2, features=['result.rating','result.user_ratings_total','price_level']):
    """Test means of a feature set of two samples
    
    Args:
        s1 (dataframe): sample 1
        s2 (dataframe): sample 2
        features (list): an array of features to test
    
    Returns:
        dict: a dictionary of t-test scores for each feature where the feature name is the key and the p-value is the value
    """
    results = {}

    # Your code here
    #dictionary comprehension to return first the feature and then the computed p-value for each feature
    results = {f:st.ttest_ind(s1[f],s2[f])[1] for f in features}
    return results
```

```python
t_test_features(centro, barajas) #Averages of rating score, price level, and number of people that have rated the restaurants is significantly different
```

```python
t_test_features(centro, salamanca) #Salamanca and centro differ on price level (higher in Salamanca) and number of ratings, higher in Centro
```

```python
t_test_features(chamberi, salamanca) #different price level, Salamanca is the most expensive area
```

```python
t_test_features(chueca_malasaña, centro) #Centro has more expensive restaurants and more reviews, than Chueca/Malasañab
```

```python
t_test_features(closed_temporarily, operational) #All differences are quite substancial
```
