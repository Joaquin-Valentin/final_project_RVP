```python
import requests
import pandas as pd
import random
import time
import json



def get_data(url):
    response = requests.get(url) 
    if response.status_code == 200:
        return response.json()
    else:
        return None

def get_df(data):
    if data:
        return pd.DataFrame(pd.json_normalize(data['results']))
    else:
        return None 
```

```python
def get_places(location):
    results = []
    next_page_token = None
    for i in range(3):
        if i == 0:
            url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={location}&radius=1500&type=restaurant&key=AIzaSyCTikq4DDzQxEd2UfsgSr-S3ImmOfAvbt8"
        else:
            if next_page_token:
                url = f"{url}&pagetoken={next_page_token}"
            response = requests.get(url)
            data = json.loads(response.text)
            results += data['results']
            if 'next_page_token' in data:
                next_page_token = data['next_page_token']
            else:
                break
        time.sleep(2)
    return pd.json_normalize(results)
```

```python
def get_places_new(locations):
    all_results = []
    for location in locations:
        results = []
        next_page_token = None
        for i in range(3):
            if i == 0:
                url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={location}&radius=600&type=restaurant&key=AIzaSyCTikq4DDzQxEd2UfsgSr-S3ImmOfAvbt8"
            else:
                if next_page_token:
                    url = f"{url}&pagetoken={next_page_token}"
                response = requests.get(url)
                data = json.loads(response.text)
                results += data['results']
                if 'next_page_token' in data:
                    next_page_token = data['next_page_token']
                else:
                    break
            time.sleep(2)
        df = pd.json_normalize(results)
        all_results.append(df)
    return pd.concat(all_results, ignore_index=True)
```

```python
import requests
import pandas as pd

def get_place_details(place_ids):
    base_url = "https://maps.googleapis.com/maps/api/place/details/json?"
    df_list = []

    for place_id in place_ids:
        url = base_url + f"place_id={place_id}&key=AIzaSyASsUGkw9ee_NHDRQj-BuBut_3CKUn_2Do"
        response = requests.get(url)
        data = response.json()

        # You can access the data returned from the API here and add it to a list of dataframes
        df_list.append(pd.json_normalize(data))

    # Concatenate all dataframes into one
    result = pd.concat(df_list, ignore_index=True)
    return result
```
