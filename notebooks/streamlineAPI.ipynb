```python
import re
import requests
import pandas as pd
import json
import time
import random
```

```python
def get_data(url):
    payload={}
    headers = {}
    response = requests.request("GET", url, headers=headers, data=payload) 
    if response.status_code == 200:
        return response.json()
    else:
        return None

def get_df(url):
    data = get_data(url)
    if data:
        return pd.DataFrame(pd.json_normalize(data['results']))
    else:
        return None

def get_multiple_dfs(base_url):
    dfs = []
    for i in range(3):
        url = base_url
        if i > 0:
            previous_df = dfs[i-1]
            if 'next_page_token' in previous_df:
                next_page_token = previous_df['next_page_token'].iloc[0]
                url += f"&pagetoken={next_page_token}"
        df = get_df(url)
        if df is not None:
            dfs.append(df)
        sleep_time = random.uniform(0.5, 2) # random sleep time between 0.5 to 2 seconds
        time.sleep(sleep_time)
    return pd.concat(dfs, axis=0, ignore_index=True)
```
